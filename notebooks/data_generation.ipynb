```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_gpu_dataset():
    # Set a starting date for the forecast
    start_date = datetime.now()

    # Product IDs expanded to 15 different models
    products = ["RTX 3070", "RTX 3080", "GTX 1660", "RX 6800", "RTX 3090",
                "GTX 1650", "RX 6900 XT", "RX 6700 XT", "RTX 2060", "GTX 1070",
                "RTX 2080", "GTX 1080 Ti", "RX 5700", "GTX 1050 Ti", "RTX 2070"]

    # Extend the number of weeks to 12 weeks for 12 dates per product
    dates = [start_date + timedelta(weeks=w) for w in range(12)]

    # Generating the dataset
    data = []
    for product in products:
        for date in dates:
            actuals = np.random.randint(100, 1000)  # Random actuals between 100 and 1000
            xgboost_pred = actuals * np.random.normal(1.0, 0.05)  # Xgboost prediction with small noise
            rf_pred = actuals * np.random.normal(1.0, 0.05)  # Random Forest prediction with small noise
            lgbm_pred = actuals * np.random.normal(1.0, 0.05)  # LightGBM prediction with small noise
            extra_tree_pred = actuals * np.random.normal(1.0, 0.05)  # Extra Tree Regressor prediction with small noise
            
            data.append({
                "Product ID": product,
                "Forecast Date": date.date(),
                "Actuals": actuals,
                "Xgboost Model": xgboost_pred,
                "Random Forest Model": rf_pred,
                "LightGBM Model": lgbm_pred,
                "Extra Tree Regressor Model": extra_tree_pred
            })

    # Creating DataFrame
    return pd.DataFrame(data)

# Generate the dataset
dataset = generate_gpu_dataset()
```

```python
file_path = '../data/gpu_forecast_data.xlsx'
dataset.to_excel(file_path, index=False, sheet_name='ML Model Forecast')
```
